```python
from bs4 import BeautifulSoup,SoupStrainer
import requests
# import httplib2
import urllib.request
import urllib.parse
import pandas as pd
import tqdm
import time
import random
```

```python
base_url = 'https://otexa.trade.gov/scripts/TQValue.exe/HTSv'
colnames = ['Source','HTS','COUNTRY','2017','2018','Year to Date 11/2018','Year to Date 11/2019','NOV. 2019','Year Ending 11/2019','Year Ending 10/2019','Year Ending 9/2019','Year Ending 11/2018']
df = pd.DataFrame(columns=colnames)
# cutoff = 200
```

```python
def get_proxy_ip() :
    proxy_url = 'https://www.proxyscan.io/home/filterresult?limit=400&page=1&ping=&SelectedAnonymity=Anonymous&SelectedAnonymity=Elite&sortPing=false&sortTime=false&sortUptime=true&selectedType=HTTP&selectedType=HTTPs'
    r = requests.get(proxy_url)
    bs = BeautifulSoup(r.text, features="lxml")
    proxy_ip = bs.find('tr').find('th').text
    print("proxy_ip : " + proxy_ip)
    proxy = {"http": proxy_ip, "https": proxy_ip}
    return proxy
```

```python
def web_scraping(df,how_many,moyr,cutoff=100000):
    proxy = get_proxy_ip()
    
    status_code = None
    
    while status_code!= 200:
        try:
            resp = requests.get(base_url, proxies = proxy)
            status_code =  resp.status_code
            print("status code " + status_code)
        except:
            pass
       

    soup = BeautifulSoup(resp.text,features="lxml")
    ls = []
    i=0
    j=0
    for link in soup.find_all('a', href=True)[:how_many]:
        i=i+1
        url = urllib.parse.urljoin(base_url, link['href'])
        title = link.text
        text = link.nextSibling
        ls.append(url)
        print('Page : ',str(i))
        print(title)
        print(text)
        print(url)
        print('')
        r = requests.get(url, proxies = proxy )
        bs=BeautifulSoup(r.text, "lxml")
        table_body=bs.find('table')
        table_body
        rows = table_body.find_all('tr')
        for row in rows:
            cols=row.find_all('td')
            cols=[x.text.strip() for x in cols]
            cols.insert(0, title)
            cols.insert(0, text)
            if len(cols) == len(df.columns):
                df.loc[len(df)] = cols
        print("number of rows " + str(df.shape[0]))
        
        nrows = df.shape[0]
        
        if nrows > cutoff:
            j=j+1                 
            out = df[(df.COUNTRY.str[0] != "_" ) & (df.COUNTRY !='WORLD')]
            out.to_csv("output_"+ moyr + "_" + str(j) + ".csv")
            colnames = ['Source','HTS','COUNTRY','2017','2018','Year to Date 11/2018','Year to Date 11/2019','NOV. 2019','Year Ending 11/2019','Year Ending 10/2019','Year Ending 9/2019','Year Ending 11/2018']
            df = pd.DataFrame(columns=colnames)
            
        else:          
            out = df[(df.COUNTRY.str[0] != "_" ) & (df.COUNTRY !='WORLD')]
            out.to_csv("output_"+moyr+".csv")
        
    return out
```

```python
web_scraping(df,how_many=30, moyr = "Dec_2019",cutoff=100000)
```
